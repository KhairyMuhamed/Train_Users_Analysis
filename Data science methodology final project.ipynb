### cluster

```python
users_clean=users_clean[users_clean.age!=0]
users_clean
```

```python
apart_of_data_set_to_clusster=pd.DataFrame()
apart_of_data_set_to_clusster['Age']=users_clean['age']
apart_of_data_set_to_clusster['Secs_elapsed']=users_clean['secs_elapsed']
apart_of_data_set_to_clusster.describe()
apart_of_data_set_to_clusster.info()
apart_of_data_set_to_clusster=apart_of_data_set_to_clusster.dropna()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5).fit(apart_of_data_set_to_clusster)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(apart_of_data_set_to_clusster['Age'], apart_of_data_set_to_clusster['Secs_elapsed'], c= kmeans.labels_.astype(float), s=50, alpha=10)
plt.scatter(centroids[:, 0], centroids[:, 1], c='green', s=50)
plt.show()
```

```python
apart_of_data_set_to_clusster=apart_of_data_set_to_clusster[apart_of_data_set_to_clusster.Age <50]
kmeans = KMeans(n_clusters=5).fit(apart_of_data_set_to_clusster)
centroids = kmeans.cluster_centers_
plt.scatter(apart_of_data_set_to_clusster['Age'], apart_of_data_set_to_clusster['Secs_elapsed'], c= kmeans.labels_.astype(float), s=50, alpha=10)
plt.scatter(centroids[:, 0], centroids[:, 1], c='green', s=50)
plt.show()
```
